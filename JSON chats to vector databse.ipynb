```python
import json

import pinecone
from langchain.vectorstores import Pinecone
from langchain.embeddings import SentenceTransformerEmbeddings

from langchain.document_loaders import DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
def load_docs(directory):
  loader = DirectoryLoader(directory)
  documents = loader.load()
  return documents
```

```python
directory = 'Path to/data/json'
```

```python
path = 'Path to/data/json/chat1.json'
```

```python
with open(path, "r") as file:
    json_data = json.load(file)
```

```python
chat = ""
for item in json_data:
    if "text" in item:
        chat += item["text"] + " "

# Print the continuous text
print(chat)
```

```python
path2 = 'Path to/data/text/textchat.txt'

# Save the text to a text file
with open(path2, "w") as file:
    file.write(chat)
```

```python
directory = 'Path to/data/text'
documents = load_docs(directory)
```

```python

def split_docs(documents,chunk_size=200,chunk_overlap=5):
  text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
  docs = text_splitter.split_documents(documents)
  return docs
```

```python
docs = split_docs(documents)

print(docs[1].page_content)
```

```python
embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
# initialize pinecone
pinecone.init(
    api_key="PINECONE KEY",
    environment="ENV NAME"
)

index_name = "slackqna"

index = Pinecone.from_documents(docs, embeddings, index_name=index_name)
```

```python
def get_similiar_docs(query,k=1,score=False):
  if score:
    similar_docs = index.similarity_search_with_score(query,k=k)
  else:
    similar_docs = index.similarity_search(query,k=k)
  return similar_docs
```

```python
query = "training a linear regression"
similar_docs = get_similiar_docs(query)
similar_docs
```

```python
query = "overall functional fit"
similar_docs = get_similiar_docs(query)
similar_docs
```
